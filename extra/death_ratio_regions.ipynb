```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn 
```

```python
data = pd.read_csv('data.csv')
data['data'] = pd.to_datetime(data['data'], dayfirst= True)
```

```python
data.set_index('data', inplace=True)
```

```python
df  = data.copy()
```

```python
df.columns
```

```python
df[df['data_dados'] == '26-01-2021 00:00'].transpose().to_csv('day_per_region.csv')
```

```python
confirmed_continental = [
'confirmados_arsnorte', 
'confirmados_arscentro', 
'confirmados_arslvt',
'confirmados_arsalentejo',
'confirmados_arsalgarve', 
]

deaths_continental = [
'obitos_arsnorte', 
'obitos_arscentro', 
'obitos_arslvt',
'obitos_arsalentejo',
'obitos_arsalgarve', 
]

recoveries_continental = [
'recuperados_arsnorte', 
'recuperados_arscentro', 
'recuperados_arslvt',
'recuperados_arsalentejo',
'recuperados_arsalgarve', 
]

zones = [
'arsnorte',
'arscentro',
'arslvt',
'arsalentejo',
'arsalgarve',  
]


variables_in_use =   confirmed_continental + deaths_continental + recoveries_continental
df = df[variables_in_use]
```

```python
df_pct_change1 = df.pct_change(periods = 1).replace(float('inf'), 0)
df_pct_change1
```

```python
df_pct_change7 = df.pct_change(periods = 7).replace(float('inf'), 0)
df_pct_change7
```

```python
df_pct_change67 = df.pct_change(periods = 67).replace(float('inf'), 0)
df_pct_change67
```

```python
fig, ax = plt.subplots(2,1, sharex=True)
df[confirmed_continental].plot(linestyle='-', ax = ax[0])
df[deaths_continental].plot(linestyle='-', ax = ax[1])
```

```python
fig, ax = plt.subplots(3,2, sharex=True, sharey=True, figsize=(40,15))

df_pct_change1[confirmed_continental].iloc[-75:-1].plot(ax = ax[0,0])#, title='Percentage change in Confirmed Cases (MA1)')
df_pct_change1[deaths_continental].iloc[-75:-1].plot(ax = ax[0,1], title='Percentage change in Covid-related Deaths (MA1)')#idxmax()


df_pct_change7[confirmed_continental].iloc[-75:-1].plot(ax = ax[1,0], title='Percentage change in Confirmed Cases (MA7)')
df_pct_change7[deaths_continental].iloc[-75:-1].plot(ax = ax[1,1], title='Percentage change in Covid-related Deaths (MA7)')#idxmax()

df_pct_change67[confirmed_continental].iloc[-75:-1].plot(ax = ax[2,0], title='Percentage change in Confirmed Cases (MA67)')
df_pct_change67[deaths_continental].iloc[-75:-1].plot(ax = ax[2,1], title='Percentage change in Covid-related Deaths (MA67)')#idxmax()

ax[0,0].tick_params(axis='x', colors='white')
ax[0,0].tick_params(axis='y', colors='white')
ax[0,0].set_title('Percentage change in Confirmed Cases (MA1)', color='w')

ax[0,1].tick_params(axis='x', colors='white')
ax[0,1].tick_params(axis='y', colors='white')
ax[0,1].set_title('Percentage change in Covid-related Deaths (MA1)', color ='w')

ax[1,0].tick_params(axis='x', colors='white')
ax[1,0].tick_params(axis='y', colors='white')
ax[1,0].set_title('Percentage change in Confirmed Cases (MA7)', color='w')


ax[1,1].tick_params(axis='x', colors='white')
ax[1,1].tick_params(axis='y', colors='white')
ax[1,1].set_title('Percentage change in Covid-related Deaths (MA7)', color ='w')

ax[2,0].tick_params(axis='x', colors='white')
ax[2,0].tick_params(axis='y', colors='white')
ax[2,0].set_title('Percentage change in Confirmed Cases (MA67)', color='w')

ax[2,1].tick_params(axis='x', colors='white')
ax[2,1].tick_params(axis='y', colors='white')
ax[2,1].set_title('Percentage change in Covid-related Deaths (MA67)', color ='w')

ax[0,0].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)
ax[0,1].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)
ax[1,0].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)
ax[1,1].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)
ax[2,0].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)
ax[2,1].legend(bbox_to_anchor=(1.0, 1.0), loc='upper left', markerscale = 0.3)



fig.savefig('deaths_region.png', dpi =300)
```
